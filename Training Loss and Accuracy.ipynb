```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
with open('data/training_history_2.pkl', 'rb') as f:
    history = pickle.load(f)
```

```python
loss = history['loss']
```

```python
sns.set()
plt.plot(loss)
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.show()
```

```python
accuracy = history['acc']
sns.set()
plt.plot(accuracy)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.show()
```
